```python
#Import libraries 
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
pd.set_option("display.max_rows",1000)
pd.set_option("display.max_columns",1000)
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import accuracy_score
```

#### 1 Load in the dataset

```python
df = pd.read_csv("data/air_satisfication.csv")
df = df.dropna()
df
```

```python
encoded_y = LabelEncoder().fit_transform(df["satisfaction"])
print(encoded_y)
print(len(encoded_y), len(df["satisfaction"]))
print(np.unique(encoded_y))
```

```python
df.columns
```

```python
df['Gender'] = LabelEncoder().fit_transform(df['Gender'])
df['Customer Type'] = LabelEncoder().fit_transform(df['Customer Type'])
df['Type of Travel'] = LabelEncoder().fit_transform(df['Type of Travel'])
df['Class'] = LabelEncoder().fit_transform(df['Class'])
```

#### 2 Train Random Forest model

```python
def get_dataset(features):
    #Filter just features
    df_features = df[features].copy()
    X = df_features.values.copy()
    #Filter genres and add label column
    df_features.loc[df['satisfaction'] == "satisfied", "label"] = 0
    df_features.loc[df['satisfaction'] == "neutral or dissatisfied", "label"] = 1
    
    X = df_features.drop(columns=['label']).values  # Features matrix
    y = df_features['label'].values  # Target labels

    return X, y

def train(dataset, rf=True):
    X, y = dataset
    # Fit the model
    model = RandomForestClassifier(oob_score=True, random_state=42, n_estimators=200, n_jobs=-1)
    model.fit(X, y)
    print("OOB accuracy", model.oob_score_)
    return model 
```

```python
# features = ["Flight Distance", "Food and drink"]
# train(get_dataset(features))
features = ['Gender', 'Customer Type', 'Age', 'Type of Travel',
       'Class', 'Flight Distance', 'Inflight wifi service',
       'Departure/Arrival time convenient', 'Ease of Online booking',
       'Gate location', 'Food and drink', 'Online boarding', 'Seat comfort',
       'Inflight entertainment', 'On-board service', 'Leg room service',
       'Baggage handling', 'Checkin service', 'Inflight service',
       'Cleanliness', 'Departure Delay in Minutes', 'Arrival Delay in Minutes']
train(get_dataset(features))
```
